```python
import pandas as pd
import re
```

```python
json_path = r"E:\DianWork\DianGPT\昇腾AI\dataset\train_rate.json"
# 读取文件中的每一行并将其作为独立的 JSON 对象
with open(json_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()

# 将每行的 JSON 对象存储在一个列表中
json_data = '[' + ','.join([line.strip() for line in lines]) + ']'

# 使用 pandas 读取 JSON 数据
df = pd.read_json(json_data)

```

```python
df['rate'] = df['rate'].apply(lambda x : int(str(x)[-1]))
```

```python
df['rate'].value_counts()
```

```python
len(df)
```

```python
df
```

```python
rate_1_samples = df[df['rate'] == 1].sample(n=50, random_state=1)
rate_2_samples = df[df['rate'] == 2].sample(n=50, random_state=1)
rate_3_samples = df[df['rate'] == 3].sample(n=50, random_state=1)
rate_4_samples = df[df['rate'] == 4].sample(n=50, random_state=1)

# 合并所有采样数据
sampled_df = pd.concat([rate_1_samples, rate_2_samples, rate_3_samples, rate_4_samples])

# 重置索引
sampled_df = sampled_df.reset_index(drop=True)
```

```python
sampled_df
```

```python
out_df = sampled_df.sample(200)
```

```python
out_df
```

```python
# 将sampled_df导出为JSON文件，每一行对应一个JSON对象
out_df.to_json(r'E:\DianWork\DianGPT\昇腾AI\dataset\testset.json', orient='records', lines=True,force_ascii=False)
```

```python
def extract_number(solution):
    """提取数学表达式中的数值"""
    return int(float(re.findall(r"[-+]?\d*\.\d+|\d+", solution)[-1]))
```

```python
solution = "平均值为 57.833333333333336"
extract_number(solution)
```
